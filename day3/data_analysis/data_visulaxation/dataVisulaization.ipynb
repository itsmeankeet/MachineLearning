```python
#figure size deda jaila plot bhnda mathi garney hai tw 
plt.figure(figsize=(15,5))
plt.plot(df1['Date'], df1['Sales_Volume'],marker = 'o', c= 'r', label= 'Sales_Volume')
plt.xlabel('Date',fontsize= 10, color= 'blue', style= 'italic')
plt.ylabel('Sales_Volume',fontsize= 10, color= 'blue', style= 'italic')
plt.title('Sales_Volume over Time', fontsize= 15)
plt.grid(True)
#grid lae box haru ma dekhauxa 
plt.legend()
plt.xticks(rotation= 45)
plt.show()
```

```python
#count plot and bar plot 
# bar plot 
''' 
kasto type ko data legney: 
compare values across different categories
we can use categoriacal data and continuous data 
 purpose:
    1. to compare quantities -> compare values across different groups
    2. also use to the distribution of a categorical data
Structure:
    x axis ma categorical data 
    y axis ma continuous data 
Example: 
    different age group ko manxey 
    college ko aadhar ma passing rate
''' 
```

```python
plt.bar(df['Product_Category'], df['Profit'], edgecolor= 'black')
plt.xlabel('Product_Category',fontsize= 10, color= 'blue', style= 'italic')
plt.ylabel('Profit',fontsize= 10, color= 'blue', style= 'italic')
plt.title('Profit by Product_Category', fontsize= 15)
plt.xticks(rotation= 45)
plt.show()

#what does edgecolor do?
#edgecolor is used to set the color of the edge of the bars.

#task edgecolor = ' black what does it do? why it is essential?
#edgecolor = ' black' is used to set the color of the edge of the bars to black.
#it is essential because it helps to make the bars more visible and distinct from the background.

#task 2: 
#task 2: 
```

```python
#count plot 
'''
'''
count = df['Product_Category'].value_counts().reset_index()
count.plot(kind = 'bar', color = 'green')
plt.xlabel('Product_Category',fontsize= 10, color= 'blue', style= 'italic')
plt.ylabel('Count',fontsize= 10, color= 'blue', style= 'italic')
plt.title('Count of Product_Category', fontsize= 15)
plt.xticks(rotation= 45)
plt.show()
```

```python
#example of pie chart
#first ma value count 
product_category = df['Product_Category'].value_counts()
plt.pie(product_category,labels= product_category.index, autopct='%1.2f%%', startangle= 50, colors= ['red', 'blue', 'green'])
plt.title('Product_Category')
plt.show()
```

```python
product_category.plot(kind = 'pie', autopct= '%1.1f%%', startangle= 50, colors= ['red', 'blue', 'green'])
plt.title('Product_Category')
plt.show()
```

```python
plt.pie(df['Region'].value_counts(), labels= df['Region'].value_counts().index, autopct= '%1.1f%%', colors= ['red', 'blue', 'green', 'grey'])
plt.title('Region Distribution')
plt.show()

```

```python
region_count = df['Region'].value_counts()
plt.figure(facecolor='grey')
plt.title('Region Distribution', fontsize= 15, color= 'black')
region_count.plot(kind = 'pie', autopct = '%1.1f%%')
```

```python
category = ['apple', 'banana', 'orange', 'mango', 'grape', 'pineapple', 'strawberry']
counts = [3,4,2,5,7,2,7]
colors = sns.color_palette('Set3', len(category))
plt.figure(facecolor='black')
#we use facecolor to set the background color of the graph
plt.pie(counts, labels=[f'{cat}: {count}' for cat, count in zip(category, counts)], autopct='%1.1f%%',startangle=0, colors=colors, pctdistance= 0.8)
plt.rcParams['text.color'] = 'green' # This will set the percentage text color to white

plt.title('Fruit Sold Per day', fontsize= 15, color= 'green')
plt.show()
```
