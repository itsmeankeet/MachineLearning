## Finding outliers using z-score 
### Formula:
    - z = (x - mean)/sd
    - x = data point 
    - mean = mean of the column
    - sd = standard deviation
    - if z score id greater than or smaller tha (3,-3) then that is outliers
### Conditions to use z-score:
    - COlums should be normally distributed

```python
#this is the example of z-score example 
x = [10,20,30,40,50,60,7,300]
mean = np.mean(x)
sd = np.std(x)
# z score for 10
z = (300 - mean)/sd
```

```python
z
```

#find all the outliers of the remaining bell-shaped coloums
- now we have 2 steps.
- 1. Triming (removing the rows)

```python
def find_ul_ll(col_name):
    ul = df[col_name].mean() + 3*df[col_name].std()
    ll = df[col_name].mean() - 3 * df[col_name].std()
    return ul, ll
```

```python
#removing outliers for skew data(IQR method is used)
Q1 = df3['GDP'].quantile(0.25)
Q3 = df3['GDP'].quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
#now we calculate the lower and upper limit 
lower_bound = Q1 - 1.5*IQR
upper_bound = Q1 + 1.5*IQR
```

```python
#triming 
df_1 = df[(df['GDP'] >= lower_bound) & (df['GDP'] <= upper_bound)] 
```

```python
df_1.shape
```

```python
def remove_outliers_iqr(column):
    global df_1
    for col in column:
        Q1 = df_1[col].quantile(0.25)
        Q3 = df_1[col].quantile(0.75)
        IQR = Q3 - Q1
        upper_bound = Q1 + 1.5*IQR
        lower_bound = Q1 -1.5*IQR

        df_1 = df_1[(df_1[col] >= lower_bound) & (df_1[col  ] <= upper_bound)] 
    return df_1
```

```python
df_2 = remove_outliers_iqr(colum)
```
