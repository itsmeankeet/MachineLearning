```python
df['Android Ver'].fillna(df['Android Ver'].mode()[0],inplace=True)
```

```python
df['Current Ver'].fillna(df['Current Ver'].mode()[0],inplace=True)

df['Content Rating'].fillna(df['Content Rating'].mode()[0],inplace=True)

df['Type'].fillna(df['Type'].mode()[0],inplace=True)

df['Rating'].fillna(df['Rating'].mean(),inplace=True)
```

```python
#barplots

sns.barplot(x='Category',y='Rating',data=df)
plt.show()
```

```python
plt.scatter(x='Category',y='Rating',data=df)
```

```python
df[df['Rating'] > 5].groupby(['Category','Rating']).count().reset_index()
```

```python
data1 = df.tail(20)
data1
```

```python
sns.barplot(x='Category',y='Rating',data=data1)
plt.xticks(rotation=70)
plt.show()
```

```python
plt.scatter(x = 'Size', y = 'Rating',data = data1)
plt.xticks(rotation=70)
plt.show()
```

```python
# The error is that plt.plot() doesn't accept x and y as keyword arguments
# Instead we need to pass the data directly from the dataframe columns
plt.plot(data1['Last Updated'], data1['Rating'], label='Size vs Rating')
plt.xticks(rotation=70)
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x='Category',data=df)
plt.xticks(rotation=70)
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.bar(df['Category'],df['Rating'])
plt.xticks(rotation=70)
plt.show()
```
